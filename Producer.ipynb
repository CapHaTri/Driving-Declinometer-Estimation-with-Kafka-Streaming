```python
import pandas as pd
from kafka import KafkaProducer
import time

import json
```

```python
KAFKA_TOPIC_NAME_CONS = "Driving_Declinometer_test"
KAFKA_BOOTSTRAP_SERVERS_CONS = 'localhost:9092'
kafka_producer_obj = KafkaProducer(bootstrap_servers=KAFKA_BOOTSTRAP_SERVERS_CONS)
```

```python
print("Kafka Producer Application Started ... ")
filepath = "test.csv"
data_df = pd.read_csv(filepath)
data_dict = data_df.to_dict(orient="records")
    
message_list = []
message = None
```

```python
for message in data_dict:
    message_fields_value_list = {}

    message_fields_value_list["Epm_nEng_100ms"] = message["Epm_nEng_100ms"]
    message_fields_value_list["VehV_v_100ms"] = message["VehV_v_100ms"]
    message_fields_value_list['ActMod_trqInr_100ms'] = message['ActMod_trqInr_100ms']
    message_fields_value_list["RngMod_trqCrSmin_100ms"] = message["RngMod_trqCrSmin_100ms"]
    message_fields_value_list["CoVeh_trqAcs_100ms"] = message["CoVeh_trqAcs_100ms"]
    message_fields_value_list["Clth_st_100ms"] = message["Clth_st_100ms"]
    message_fields_value_list["CoEng_st_100ms"] = message["CoEng_st_100ms"]
    message_fields_value_list["Com_rTSC1VRVCURtdrTq_100ms"] = message["Com_rTSC1VRVCURtdrTq_100ms"]
    message_fields_value_list["Com_rTSC1VRRDTrqReq_100ms"] = message["Com_rTSC1VRRDTrqReq_100ms"]

    json_message = json.dumps(message_fields_value_list)

    print("Message Type: ", type(json_message))
    print("Message: ", json_message)
       
    kafka_producer_obj.send(KAFKA_TOPIC_NAME_CONS, json_message.encode('utf-8'))
    time.sleep(5)

print("Kafka Producer Application Completed. ")
```
